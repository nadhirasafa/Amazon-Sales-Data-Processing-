## Soal Nomor 1

```python
query1 = """select category, avg(discounted_price) as avg_discount
from amazon_sales
group by 1
order by 2 desc
limit 5"""
pd.read_sql_query(query1,conn)
```

## Soal Nomor 2

```python
query2 = """
SELECT COUNT(distinct user_id) as jumlah_review, category
FROM amazon_sales
WHERE review_title LIKE '%Good Product%' AND
category IN ('Computers & Accessories', 'Office Products')
group by category
"""
pd.read_sql_query(query2,conn)
```

## Soal Nomor 3

```python
query3 = """ SELECT (actual_price-discounted_price)*qty as total_sales,
strftime('%m',order_date) as month,
strftime('%Y',order_date) as year
FROM amazon_sales
WHERE strftime('%Y', order_date) = '2004' AND
category NOT IN ('Office Products', 'Home & Kitchen')
GROUP BY month
ORDER BY total_sales ASC
limit 5;
"""
pd.read_sql(query3,conn)
```

## Soal Nomor 4

```python
query4 = """SELECT category, arrival_date, order_date,
abs(round(avg(julianday(DATE(order_date)) - julianday(DATE(arrival_date))))) AS lama_pengiriman
FROM amazon_sales
WHERE category NOT IN ('Car & Motorbike', 'Toys & Games')
GROUP BY category
ORDER BY lama_pengiriman ASC;"""
pd.read_sql(query4,conn)
```

## Soal Nomor 5

```python
query5= """
SELECT SUM(discounted_price) AS total_sales,
COUNT(*) AS total_order,
SUM(discounted_price) / COUNT(*) AS aov,
strftime('%Y', order_date) AS year
FROM amazon_sales
WHERE strftime('%d', order_date) BETWEEN '01' AND '15'
GROUP BY year;"""
pd.read_sql(query5,conn)
```
